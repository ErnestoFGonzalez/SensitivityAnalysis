```python
import numpy as np
import pandas as pd
import pylab
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib notebook
```

Let's start by creating some dataframe to test our PCA-based algorithm for sensitivity analysis.

```python
x = np.linspace(0,10,1000)
y = np.random.randint(low=0, high=10, size=1000)

# output - depends only on x
t = x**2 + 34*x - 81

df = pd.DataFrame({'x':x, 'y':y, 't':t})

df

figure = pylab.figure().gca(projection='3d')
figure.scatter(df['x'], df['y'], df['t'])
figure.set_xlabel('x')
figure.set_ylabel('y')
figure.set_zlabel('t')
pylab.show()
```
