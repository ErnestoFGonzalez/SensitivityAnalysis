# Column to output influence
In this file we will plot output data trajectories in plane (`Column i`,`Column 7`), with $i=[6]$, for our dataset `dados.dat`.

```python
import numpy as np
import pandas as pd
import pylab
from mpl_toolkits.mplot3d import Axes3D
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
%matplotlib notebook
```

Let's start by reading the data file `dados.dat` and creating a dataframe.

```python
with open('dados.dat', 'r') as infile:
    readlines = infile.readlines()
    
    df = []

    for line in readlines: 
        temp = line.split('   ')
        for elem in temp:
            if elem == '':
                temp.remove(elem)
                continue
            temp[temp.index(elem)] = float(elem.strip())
        df.append([temp[0], float(temp[1]), temp[2], 
                   temp[3], temp[4], temp[5], temp[6], temp[7]])
            
df = pd.DataFrame(df)
```

```python
df
```

## Plane (Column 0, Output)

```python
x = df.loc[:, [0]].values
y = df.loc[:, [7]].values

# plot
pylab.plot(x, y, 'bo')
```

To try and see better the behaviour for low values of `Output` let's exclude rows where `Output`$>50$.

```python
# filter for rows with Column 7 value lesser (or equal) than 50
filtered_y_idx = df.index[df[7]<=50]
x = df.loc[filtered_y_idx, [0]].values
y = df.loc[filtered_y_idx, [7]].values

# plot
pylab.plot(x, y, 'bo')
```
